```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("2019.csv")
og = data
data
```

```python
info = pd.read_csv("pop_density.csv", decimal=",")
info
```

```python
## analyze population density
```

```python
pops = info[["Country", "Pop. Density (per sq. mi.)"]]
pops
```

```python
pops["pop_density"] = pops["Pop. Density (per sq. mi.)"]
pops = pops.drop("Pop. Density (per sq. mi.)",axis=1)
pops
```

```python
## Clean and prepare data
```

```python
pops["Country"] = pops["Country"].str.rstrip()
pops
```

```python
pops[pops["Country"] == "Afghanistan"]
```

```python
data["Country"] = data["Country or region"]
data = data.drop("Country or region",axis=1)
data
```

```python
s = set(data["Country"])
print("Afghanistan" in s)
```

```python
ls = []
for country in data["Country"]:
    s = set(pops["Country"])
    if country in s:
        r = pops[pops["Country"] == country]
        ls.append(r["pop_density"].values[0])
    else:
        ls.append('')
ls = np.array(ls)
len(ls)
```

```python
data["pop_density"] = ls
data["pop_density"] = data["pop_density"].replace('',np.nan)
data
```

```python
data = data.dropna()
data
```

```python
## linreg on OG data
```

```python
og = og.drop(["Country or region", "Country"],axis=1)
```

```python
from sklearn.linear_model import LinearRegression

m = LinearRegression()
```

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(og, test_size=.2, random_state=42)
```

```python
X_train = train_set.drop(["Overall rank", "Score"], axis=1)
X_test = test_set.drop(["Overall rank", "Score"], axis=1)
X_train
```

```python
y_train = train_set["Score"]
y_test = test_set["Score"]
```

```python
m.fit(X_train, y_train)
```

```python
m.score(X_train, y_train)
```

```python
m.score(X_test, y_test)
```

```python
## now perform linreg on updated dataset with population density factor

train_set, test_set = train_test_split(data, test_size=.2, random_state=42)

X_train = train_set.drop(["Overall rank", "Score", "Country"], axis=1)
X_test = test_set.drop(["Overall rank", "Score", "Country"], axis=1)

y_train = train_set["Score"]
y_test = test_set["Score"]

m.fit(X_train, y_train)
```

```python
data
```

```python
m.score(X_train, y_train)
```

```python
m.score(X_test, y_test)
```
